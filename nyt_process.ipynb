```python
# climate+change
# gun+control
# medicare+for+all
# brexit
# north+korea
# fake+news
# illegal+immigration
import pandas as pd
import urllib3
from bs4 import BeautifulSoup
```

```python
#1 change file name here based on sub-topic
df = pd.read_csv("input/raw/nyt_guncontrol.csv")
```

```python
df.head()
```

```python
df = df["web_url"]
```

```python
http = urllib3.PoolManager()

text = ""
for url in df:
    
    response = http.request('GET', url)
    
    soup = BeautifulSoup(response.data, 'html.parser')
    
    matches = soup.select("main#site-content article#story section div.StoryBodyCompanionColumn > div > p")
    for item in matches:
        text += item.get_text()+"\n"
```

```python
#2 change file name here based on sub-topic
file = open("./input/mr/nyt_guncontrol.txt", "w", encoding="utf-8") 
file.write(text)
file.close()
```
